###**Speed** 
now let's see how is it faster than regular for loop and qwack maths?

```python
# let's first create a pandas series:
import pandas as pd
s = pd.Series([100.00, 120.00, 101.00, 3.00])
s
```

```python
# to sum the array we can use the usual for loop
total = 0
for item in s:
    total+=item
print(total)
```

```python
# and we can also use Vectorization, in one line:

total = np.sum(s)
print(total)
```

```python
# you can also use pandas quick methods 
s.sum()
```

```python
#this creates a big series of random numbers with  1000  numbers
s = pd.Series(np.random.randint(0,1000,10000))
# let's see the length of the series

len(s)
```

```python
%%timeit -n 100
# let's sum the whole thing using for loop
summary = 0
for item in s:
    summary+=item
```

```python
%%timeit -n 100
# let's sum the whole thing using Vectorization
summary = np.sum(s)
```

```python
%%timeit -n 100
# we can also see how pandas method is doing
s.sum()
# for you to know that pandas methods are build in with numpy, but not all numpy functions work on it
# so you would better use numpy functions everytime
```

```python
# ler's create a numpy array, and you can call it matrix as it's consists of 3,4
s = np.array([[100.00, 120.00, 101.00, 3.00],
             [100.00, 120.00, 101.00, 3.00],
             [100.00, 120.00, 101.00, 3.00]])
```

```python
# let's print the natrux
print(f'array =\n{s} \n\nshape={s.shape}')
print(type(s))
```

```python
# let's sum the whole matrix

s.sum()


```

```python
# lets sum the matrix on the column axis
# notice it returns 3 values one for each row
s.sum(axis=1)
```

```python
# lets sum the matrix on the rows axis
# notice it returns 4 values one for each column
s.sum(axis=0)
```
